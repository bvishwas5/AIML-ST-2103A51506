```python
import random
import math as mt
x=random.uniform(-10,10)
y=random.uniform(-10,10)
eta=0.01
max_iter=10000
```

```python
def gradient(x,y):
  return 6*x,10*y
```

```python
for iter in range(max_iter):
  del_x,del_y=gradient(x,y)
  x=x-eta*del_x
  y=y-eta*del_y
print('optimal point is =',x,y)
print('objective function =', 3*mt.pow(x,2)+5*mt.pow(y,2)+10)
```
