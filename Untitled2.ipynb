```python
# Data
X = [0.1, 0.2, 0.3, 0.4]
Y = [2.2, 2.4, 2.6, 2.8]

# Initial parameters
m = 2
c = -2

# Learning rate
lr = 0.1

# Function to compute gradients and update m, c
def update_parameters(X, Y, m, c, lr):
    n = len(X)
    dm = 0
    dc = 0

    for i in range(n):
        y_pred = m * X[i] + c
        error = y_pred - Y[i]
        dm += error * X[i]
        dc += error

    # Average gradients
    dm /= n
    dc /= n

    # Gradient Descent update
    m = m - lr * dm
    c = c - lr * dc

    return m, c

# Perform 2 iterations
print("Initial m = {}, c = {}".format(m, c))
for epoch in range(2):
    m, c = update_parameters(X, Y, m, c, lr)
    print("After iteration {}: m = {:.4f}, c = {:.4f}".format(epoch + 1, m, c))
```
