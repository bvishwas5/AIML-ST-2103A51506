```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Input data
data = np.array([
    [0.1, 0.2],
    [0.2, 0.3],
    [0.3, 0.4],
    [-1.0, -2.0],
    [-3.0, -4.0],
    [-4.0, -5.0]
])

# Step 2: Create KMeans model with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(data)

# Step 3: Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Step 4: Display results
print("Cluster Labels for each point:")
for point, label in zip(data, labels):
    print(f"Point {point} is in Cluster {label}")

print("\nCluster Centroids:")
print(centroids)

# Step 5: Plotting the clusters
colors = ['red' if label == 0 else 'blue' for label in labels]

plt.scatter(data[:, 0], data[:, 1], c=colors, label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='black', label='Centroids')
plt.title("K-Means Clustering (k=2)")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.grid(True)
plt.show()
```
