```python
import numpy as np

# Define the function and gradient
def f(x, y):
    return 3*x**2 + 5*np.exp(-y) + 10

def gradient(x, y):
    df_dx = 6 * x
    df_dy = -5 * np.exp(-y)
    return np.array([df_dx, df_dy])

# Gradient Descent parameters
alpha = 0.1  # learning rate
epsilon = 1e-6
max_iters = 1000
x, y = 1.0, 1.0  # initial guess

for i in range(max_iters):
    grad = gradient(x, y)
    new_x = x - alpha * grad[0]
    new_y = y - alpha * grad[1]

    # Check for convergence
    if np.linalg.norm([new_x - x, new_y - y]) < epsilon:
        break
    x, y = new_x, new_y

print(f"Optimal point: ({x:.5f}, {y:.5f})")
print(f"Minimum value: {f(x, y):.5f}")
```
