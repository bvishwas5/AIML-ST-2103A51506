```python
import random
import math as mt
x=random.uniform(-10,10)
eta=0.01
max_iter=10000
```

```python
def gradient(x):
  return 2*x
```

```python
for iter in range(max_iter):
  del_x=-eta*gradient(x)
  x=x+del_x
print('optimal point is =', x)
print('objective funtion =', mt.pow(x,2)+5)
```
